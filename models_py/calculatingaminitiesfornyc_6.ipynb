```python
import requests
import pandas as pd
import geopy.distance
from geopy.geocoders import Nominatim
import time
from google.colab import drive
```

```python
geolocator = Nominatim(user_agent="osm_data_extractor")
```

```python
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/sample_files/armslengthsales_2022_valid_20230404.csv')
df
```

```python
df = df.loc[6000:7000]
df
```

```python
data_list = []
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
```

```python
radius = 1000
```

```python
categories = {
    "Tsp": '["highway"]',
    "Atr": '["tourism"]',
    "Edc": '["amenity"="school"]',
    "Hth": '["amenity"="hospital"]',
    "Rst": '["amenity"="restaurant"]',
    "Rtl": '["shop"]'
}
```

```python
def fetch_osm_data(lat, lon, category, tag):
    query = f"""
    [out:json];
    (
      node{tag}(around:{radius},{lat},{lon});
    );
    out center;
    """

    response = requests.get(overpass_url, params={'data': query})
    if response.status_code == 200:
        json_data = response.json()
        nodes = json_data.get("elements", [])

        count = len(nodes)
        distances = [
            geopy.distance.distance((lat, lon), (node["lat"], node["lon"])).meters
            for node in nodes
        ]
        avg_distance = sum(distances) / count if count > 0 else 0

        return count, avg_distance
    else:
        print(f"Failed to fetch data for {category}")
        return 0, 0
```

```python
for index, row in df.iterrows():
    address = row["Address"]

    try:
        # Convert address to latitude and longitude
        location = geolocator.geocode(address, timeout=10)
        if location:
            lat, lon = location.latitude, location.longitude
            print(f"Geocoded: {address} -> ({lat}, {lon})")
        else:
            print(f"Could not geocode address: {address}")
            continue

        # Extract data for each category
        result = {"Address": address, "Latitude": lat, "Longitude": lon}
        for cat, tag in categories.items():
            count, avg_dist = fetch_osm_data(lat, lon, cat, tag)
            result[f"{cat}Num"] = count
            result[f"{cat}Dst"] = avg_dist

        # Append result to data list
        data_list.append(result)

        # To avoid API rate limiting, add a small delay
        time.sleep(1)

    except Exception as e:
        print(f"Error processing address {address}: {e}")
```

```python
data_list
```

```python
output_df = pd.DataFrame(data_list)
output_df.to_csv("osm_data_with_addresses_6.csv", index=False)

print("Data saved to osm_data_with_addresses_6.csv")
```
